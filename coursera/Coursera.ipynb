```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 
                  'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
for i in selected_words:
    products[i] = products['word_counts'].apply(lambda x: x[i] if i in x else 0 )
```

```python
products.head()
```

```python
temp = []
for i in selected_words:
    temp.append(products[i].sum())
print temp
```

```python
max(temp)
```

```python
#ignore all 3* reviews
products = products[products['rating'] != 3]
```

```python
#positive sentiment = 4* or 5* reviews
products['sentiment'] = products['rating'] >=4
```

```python
products.head()
```

```python
train,test = products.random_split(0.8,seed =0)
```

```python
model = graphlab.logistic_classifier.create(train,target='sentiment',features=selected_words,validation_set=test)
```

```python
model['coefficients']
```

```python
model['coefficients']['value'].sort()
```

```python
model.evaluate(test)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train,
                                                     target='sentiment',
                                                     features=['word_counts'],
                                                     validation_set=test)
```

```python
sentiment_model.evaluate(test)
```

```python
graphlab.canvas.set_target('ipynb')
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')
```

```python
image_train['label'].sketch_summary()
```

```python
dog = image_train[image_train['label']=='dog']
```

```python
cat = image_train[image_train['label']=='cat']
```

```python
bird = image_train[image_train['label']=='bird']
```

```python
automobile = image_train[image_train['label']=='automobile']
```

```python
dog_model = graphlab.nearest_neighbors.create(dog,features=['deep_features'],
                                             label='id')
```

```python
cat_model = graphlab.nearest_neighbors.create(cat,features=['deep_features'],
                                             label='id')
```

```python
bird_model = graphlab.nearest_neighbors.create(bird,features=['deep_features'],
                                          label='id')
```

```python
automobile_model = graphlab.nearest_neighbors.create(automobile,features=['deep_features'],
                                             label='id')
```

```python
target = image_test[0:1]
```

```python
cat_model.query(target)[0:5]['distance'].mean()
```

```python
dog_model.query(target)[0:5]['distance'].mean()
```

```python
image_test_cat = image_test[image_test['label']=='cat']
```

```python
dog_dog_neighbors = dog_model.query(image_test_dog,k=1)
```

```python

dog_distance = graphlab.SFrame({
        'dog-dog': dog_dog_neighbors['distance'],
        'dog-cat': dog_cat_neighbors['distance'],
        'dog-bird':dog_bird_neighbors['distance'],
        'dog-automobile' : dog_automobile_neigbors['distance']
    })
```

```python
dog_distance.head()
```

```python
def is_dog_correct(row):
    if row > dog_distance['dog-dog']:
        return 1
    else:
        return 0
```

```python
dog_distance['dog-cat-correct'] = dog_distance['dog-cat'] > dog_distance['dog-dog']
```
